### **Checking Null Values**

```python
def null_values(df, name):
  nulls = df.isnull().sum()
  nulls = nulls[nulls>0]
  print(f"Null values in {name}:" )
  print(nulls)

  print("-"*80)

null_values(defects, "defects")
null_values(modules, "modules")
null_values(sprint_module_summary, "sprint_module_summary")
null_values(sprints, "sprints")
null_values(test_execution, "test_execution")
```

#### Nulls in **'days to fix'** column are retained because they indicate **open and unresolved(leaked) defects**. So no imputation or remove is needed.

### **Checking duplicate values**

```python
def duplicate_values(df, name):
    duplicates = df.duplicated().sum()
    print(f"Duplicate rows in {name}: {duplicates}")
    print("-"*80)


duplicate_values(defects, "defects")
duplicate_values(modules, "modules")
duplicate_values(sprint_module_summary, "sprint_module_summary")
duplicate_values(sprints, "sprints")
duplicate_values(test_execution, "test_execution")
```
